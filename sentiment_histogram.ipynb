```r
csv <- read.csv("hn_sentiments.csv", header=TRUE, row.names=NULL, stringsAsFactors=FALSE)
str(csv)
n_comments <- as.numeric(csv$comments)
sums <- as.numeric(csv$sum)
means <- as.numeric(csv$mean)
log_means <- means * log(n_comments)
```

```r
hist(log(n_comments))
```

```r
hist(means, main="Means", breaks=50, xlim=c(-1,1), labels=T, freq=T)
```

```r
hist(sums, main="Sums", breaks=50, labels=T, freq=T)
```

```r
plot(n_comments,means)
```

```r
plot(n_comments,log_means)
hist(log_means)
# deviations <- log_means / dev
# hist(deviations)
```

```r
dev <- sd(log_means, na.rm = T)
avg = mean(log_mean, na.rm=T)
str(avg)
str(dev)
pnorm(log_mean[1], mean=avg, sd=dev)
#normalized <- apply(log_mean, 1, function(x) pnorm(x, mean=avg, sd=dev))
normalized <- sapply(log_mean, function(x) {
    if (x >= avg || is.na(x)) {
        n <- pnorm(x, mean=avg, sd=dev) - 0.5
    } else {
        n <- 0.5 - pnorm(x, mean=avg, sd=dev, lower.tail=FALSE)
    }
    return(n)
})
hist(normalized)
```
